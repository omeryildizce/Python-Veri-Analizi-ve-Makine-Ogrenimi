```python
import sqlite3
```

```python
def selectOperasyonlari():
    connection = sqlite3.connect("chinook.db")
    cursor = connection.execute("""
        select FirstName, LastName,Country from Customers
        where Country like "%r%"
    
    """)

    for row in cursor:
        print("Firstname: {:12}".format(row[0]) , end="")
        print("Lastname: {:12}".format(row[1]), end="")
        print("Lastname: {}".format(row[2]))

    connection.close()
```

```python
def insertCustomer():
    connection = sqlite3.connect("chinook.db")
    connection.execute("""
        insert into customers (FirstName, LastName, Country, Email) 
        values ("Hüseyin", "Sahin","Turkiye", "sahin@mail.com")
    """)
    connection.commit()
    connection.close()
```

```python
def joinOperasyonlari():
    connection = sqlite3.connect("chinook.db")
    cursor = connection.execute("""
        select albums.Title, artists.Name FROM artists INNER JOIN albums
        on artists.ArtistId = albums.ArtistId
    
    """)

    for row in cursor:
        print("Name: {:10}".format(row[1]), end="\n") 
        print("Title: {}".format(row[0]) , end="\n")
        print(50*"*")

    connection.close()
```
