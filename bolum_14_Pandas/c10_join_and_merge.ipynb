```python
data1 = {
    "id":[1,2,3,4],
    "isim":["Kadir","Ömer","Ahmet","Mehmet"],
    "soyisim":["Kaya","YILDIZ","Farklı","Çam"]
}
data2 = {
    "id":[1,3,4,7],
    "isim":["Ece","Burcu","Hande","Pelin"],
    "soyisim":["Kaya","YILDIZ","Farklı","Çam"]
}
```

```python
import pandas as pd
```

```python
data1_df = pd.DataFrame(data=data1)
data2_df = pd.DataFrame(data=data2)
```

```python
data1_df
```

```python
data2_df
```

```python
pd.merge(data1_df, data2_df, on="id", how="inner")
```

```python
pd.merge(data1_df, data2_df, on="id", how="left")
```

```python
pd.merge(data1_df, data2_df, on="id", how="right")
```

```python
pd.concat([data1_df, data2_df], axis=1)
```

```python
pd.concat([data1_df, data2_df], axis=0)
```
